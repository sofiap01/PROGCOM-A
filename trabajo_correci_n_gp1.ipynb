```python
import random
import ipywidgets as widgets
from IPython.display import display, clear_output

# Opciones del juego
opciones = ["Piedra", "Papel", "Tijera"]

# Variables globales
intentosMaximos = 5
intentosRestantes = intentosMaximos
victoriasUsuario = 0
victoriasComputadora = 0
empates = 0

# Etiqueta para resultados
resultadoLabel = widgets.Label(value=f"Tienes {intentosRestantes} intentos. Elige tus 2 jugadas:")

# FunciÃ³n que evalÃºa una ronda 2 vs 2
def jugar(eleccionUsuario):
    global intentosRestantes, victoriasUsuario, victoriasComputadora, empates
    clear_output(wait=True)

    if intentosRestantes <= 0:
        print("Juego terminado. Ya no tienes mÃ¡s intentos.")
        return

    # Usuario elige dos veces
    jugadasUsuario = [eleccionUsuario, random.choice(opciones)]
    # Computadora elige dos veces
    jugadasComputadora = [random.choice(opciones), random.choice(opciones)]

    print(f"Tus jugadas: {jugadasUsuario}")
    print(f"Jugadas de la computadora: {jugadasComputadora}")

    # Comparar las 2 jugadas
    puntosUsuario = 0
    puntosComputadora = 0

    for j1, j2 in zip(jugadasUsuario, jugadasComputadora):
        if j1 == j2:
            empates += 1
        elif (j1 == "Piedra" and j2 == "Tijera") or \
             (j1 == "Papel" and j2 == "Piedra") or \
             (j1 == "Tijera" and j2 == "Papel"):
            puntosUsuario += 1
        else:
            puntosComputadora += 1

    # Decidir resultado de la ronda
    if puntosUsuario > puntosComputadora:
        print("Ganaste esta ronda ðŸŽ‰")
        victoriasUsuario += 1
    elif puntosComputadora > puntosUsuario:
        print("Perdiste esta ronda ðŸ˜¢")
        victoriasComputadora += 1
    else:
        print("La ronda fue empate ðŸ¤")
        empates += 1

    # Reducir intentos
    intentosRestantes -= 1
    if intentosRestantes > 0:
        display(widgets.Label(value=f"Te quedan {intentosRestantes} intentos."))
        display(resultadoLabel, botonPiedra, botonPapel, botonTijera)
    else:
        print("Se acabaron tus intentos. Resultados finales:")
        print(f"Victorias tuyas: {victoriasUsuario}")
        print(f"Victorias de la computadora: {victoriasComputadora}")
        print(f"Empates: {empates}")

# Crear botones
botonPiedra = widgets.Button(description="Piedra", layout=widgets.Layout(width="150px", height="80px"))
botonPapel = widgets.Button(description="Papel", layout=widgets.Layout(width="150px", height="80px"))
botonTijera = widgets.Button(description="Tijera", layout=widgets.Layout(width="150px", height="80px"))

# Conectar botones con la funciÃ³n
botonPiedra.on_click(lambda x: jugar("Piedra"))
botonPapel.on_click(lambda x: jugar("Papel"))
botonTijera.on_click(lambda x: jugar("Tijera"))

# Mostrar la interfaz inicial
display(resultadoLabel, botonPiedra, botonPapel, botonTijera)
```
